```python
import numpy as np
import matplotlib.pyplot as plt
```

# Original Model

```python
N = 30
C = np.linspace(-3, 3, N) ** 3
```

```python
x01 = np.linspace(0, 10, N) + 2*C
x02 = x01 + 2 * x01 - C
y0 = x01**2 + x02 + C
```

```python
plt.plot(x01, y0, 'k.')
plt.xlabel("x1")
plt.ylabel("y")
plt.show()
```

```python
plt.plot(x02, y0, 'k.')
plt.xlabel("x2")
plt.ylabel("y")
plt.show()
```

```python
plt.plot(x01, x02, 'k.')
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

So, there is likely a path x1 -- x2 or a counfounder of the two or a confounder of y or both.

# do(x2 = a)

```python
a = 35
```

```python
x11 = x01
x12 = [a] * N
y1 = x11**2 + x12 + C
```

```python
plt.plot(x01, y1, 'r.', label = 'do(x2 = a)')

plt.legend()
plt.xlabel("x1")
plt.ylabel("y")
plt.show()
```

Since x1 ~ y, the path direction is likely x1 --> x2, not confirmed yet.

# do(x1 = a)

```python
a = 10
x21 = a * np.ones(N)
x22 = x21 + 2 * x21 - C
y2 = x21**2 + x22 + C
```

```python
plt.plot(x02, y2, 'g.', label = 'do(x1 = a)')

plt.legend()
plt.xlabel("x2")
plt.ylabel("y")
plt.show()
```

Since y || x2, there is a direct path x1 --> x2 or there is a common confounder of all the three. :(
