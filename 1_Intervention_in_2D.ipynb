```python
import numpy as np
import matplotlib.pyplot as plt
```

# Original Model

```python
N = 30
```

```python
x01 = np.linspace(0, 10, N)
x02 = x01 + 2 * x01
y0 = x01**2 + x02
```

```python
plt.plot(x01, y0, 'k.')
plt.xlabel("x1")
plt.ylabel("y")
plt.show()
```

```python
plt.plot(x02, y0, 'k.')
plt.xlabel("x2")
plt.ylabel("y")
plt.show()
```

```python
plt.plot(x01, x02, 'k.')
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

So, there is a path x1 -- x2

# do(x2 = a)

```python
a = 35
```

```python
x11 = x01
x12 = [a] * N
y1 = x11**2 + x12
```

```python
plt.plot(x01, y1, 'r.', label = 'do(x2 = a)')

plt.legend()
plt.xlabel("x1")
plt.ylabel("y")
plt.show()
```

Since x1 ~ y, the path direction is x1 --> x2

# do(x1 = a)

```python
a = 15
x21 = a * np.ones(N)
x22 = x21 + 2 * x21 
y2 = x21**2 + x22
```

```python
plt.plot(x02, y2, 'g.', label = 'do(x1 = a)')

plt.legend()
plt.xlabel("x2")
plt.ylabel("y")
plt.show()
```

Since y || x2, the path is x1 --> x2
